```python
# 有关列表的定义
bicycles = ['trek', 'cannondale', 'redline', 'specialized'] 
print(bicycles)
```

访问列表元素的方式

```python
print(bicycles[0].title())
print(bicycles[-1])
print(bicycles[:])
print(bicycles[0:-1])
```

```python
message = f"My first bicycle was a {bicycles[0].title()}."
print(message)
```
